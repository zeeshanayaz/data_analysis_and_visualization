```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("employee.xlsx")
```

```python
designation_frequency= data.Designation.value_counts()
```

```python
designation_frequency
```

## Bar Chart

```python
designation_frequency.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Designation')
plt.ylabel('Count')
plt.title('Designation Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
```

### Pie Chart

```python
plt.figure(figsize=(6, 6))
plt.pie(
    designation_frequency,
    labels=designation_frequency.index,
    autopct='%1.1f%%',   # show percentages
    startangle=140,      # rotate start
    colors=plt.cm.Paired.colors  # nice color palette
)

plt.title('Designation Distribution')
plt.tight_layout()
plt.show()
```

### Pareto Diagram

```python
designation_frequency.cumsum()
```

```python
designation_frequency.sum()
```

```python
# Step 1: Sort values in descending order
sorted_counts = designation_frequency.sort_values(ascending=False)

# Step 2: Compute cumulative percentage
cumulative_percent = sorted_counts.cumsum() / sorted_counts.sum() * 100

# Step 3: Create the plot
fig, ax1 = plt.subplots(figsize=(8, 6))

# Bar plot
sorted_counts.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_ylabel('Count')
ax1.set_xlabel('Designation')
ax1.set_title('Pareto Diagram - Designation Distribution')

# Line plot (cumulative %)
ax2 = ax1.twinx()
cumulative_percent.plot(marker='o', color='red', ax=ax2)
ax2.set_ylabel('Cumulative Percentage (%)')
ax2.set_ylim(0, 110)

# Horizontal line at 80%
ax2.axhline(80, color='gray', linestyle='--', linewidth=1)

# Improve layout
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
