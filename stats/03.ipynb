```python
b = {
    'AgeGroup' : ['18-25', '25-35', '35-45', '45-55', '55-65', '65+'],
    'Employed' : [60, 85, 95, 97, 97, 100],
}

data = pd.DataFrame(b)
```

```python
data
```

```python
data['UnEmployed'] = 100 - data.Employed
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
bar_width = 0.35
x = range(len(data))

# Create barplot
plt.figure(figsize=(8, 4))
plt.bar(x, data['Employed'], width=bar_width, label='Employed', color='skyblue')
plt.bar([i + bar_width for i in x], data['UnEmployed'], width=bar_width, label='UnEmployed', color='salmon')

# Set x-axis labels and ticks
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.title('Employment Status by Age Group')
plt.xticks([i + bar_width / 2 for i in x], data['AgeGroup'])

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
bar_width = 0.35
gap = 0.05  # space between bars in a group
x = range(len(data))

# Create barplot
plt.figure(figsize=(8, 4))
#plt.bar(x, data['Employed'], width=bar_width, label='Employed', color='skyblue')
#plt.bar([i + bar_width for i in x], data['UnEmployed'], width=bar_width, label='UnEmployed', color='salmon')
plt.bar([i - bar_width/2 - gap/2 for i in x], data['Employed'], width=bar_width, label='Employed', color='skyblue')
plt.bar([i + bar_width/2 + gap/2 for i in x], data['UnEmployed'], width=bar_width, label='UnEmployed', color='salmon')

# Set x-axis labels and ticks
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.title('Employment Status by Age Group')
plt.xticks([i + bar_width / 2 for i in x], data['AgeGroup'])

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
xpos = np.arange(len(data['AgeGroup']))

plt.bar(xpos - 0.2, data.Employed, width=0.4, label="Employed", color='navy')
plt.bar(xpos + 0.2, data.UnEmployed, width=0.4, label="UnEmployed", color='orange')

plt.title("Employment by Age Group")
plt.ylabel("Number of People")
plt.xticks(xpos, data.AgeGroup) 
plt.legend(shadow=True)
plt.tight_layout()
plt.show()
```
