```python
chest_pain = healthcare[healthcare['DRG Definition'] == '313 - CHEST PAIN']
```

```python
alabama_chest_pain = chest_pain[chest_pain['Provider State'] == "AL"]
```

```python
alabama_chest_pain
```

```python
alabama_chest_pain.columns
```

```python
costs = alabama_chest_pain[' Average Covered Charges '].values
```

```python
costs
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(costs)
plt.title("Boxplot of Average Covered Charges")
plt.ylabel("Cost ($)")
plt.show()
```

```python
states = chest_pain['Provider State'].values
```

```python
states
```

```python
datasets = []
for state in states:
    datasets.append(chest_pain[chest_pain['Provider State'] == state][' Average Covered Charges '].values)

datasets
```

```python
len(datasets)
```

```python
plt.boxplot(datasets, label=states)
plt.figure(figsize=(20,6))
plt.title("Boxplot of Average Covered Charges")
plt.ylabel("State")
plt.show()
```
