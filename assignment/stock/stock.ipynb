```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Stock.xlsx")
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(data['Apple (AAPL)'], data['Alphabet (GOOGL)'], color='blue', alpha=0.7)

plt.xlabel('Apple (AAPL) Stock Price')
plt.ylabel('Alphabet (GOOGL) Stock Price')
plt.title('Scatter Plot: Apple vs Alphabet Stock Prices')

# Set axis limits from 0 to 1200
# plt.xlim(0, 180)
# plt.ylim(0, 1200)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(data['Apple (AAPL)'], data['Bank of America (BAC)'], color='navy', alpha=0.7)

plt.xlabel('Apple (AAPL) Stock Price')
plt.ylabel('Bank of America (BAC) Stock Price')
plt.title('Scatter Plot: Apple vs Bank of America Stock Prices')

# Set axis limits from 0 to 1200
# plt.xlim(0, 180)
# plt.ylim(0, 1200)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
correlation = data['Apple (AAPL)'].corr(data['Alphabet (GOOGL)'])
print(f"Correlation between Apple and Alphabet: {correlation:.4f}")
```

```python
correlation = data['Apple (AAPL)'].corr(data['Bank of America (BAC)'])
print(f"Correlation between Apple and Bank of America: {correlation:.4f}")
```
