```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
states0 = pd.read_csv('states0.csv')
states0
```

```python
states0.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
states0.info()
```

```python
states0.Hispanic = states0.Hispanic.str[:-1]
```

```python
states0
```

```python
states0.White = states0.White.str[:-1]
states0.Black = states0.Black.str[:-1]
states0.Native = states0.Native.str[:-1]
states0.Asian = states0.Asian.str[:-1]
states0.Pacific = states0.Pacific.str[:-1]
```

```python
states0.Hispanic = states0.Hispanic.astype(float) * 100
```

```python
states0.White = states0.White.astype(float) * 100
states0.Black = states0.Black.astype(float) * 100
states0.Native = states0.Native.astype(float) * 100
states0.Asian = states0.Asian.astype(float) * 100
states0.Pacific = states0.Pacific.astype(float) * 100
```

```python
states0
```

```python
states0.Income = states0.Income.str.replace("$", "").astype(float)
```

```python
#states0.GenderPop.str.split("_")

# states0["Males"], states0["Females"] = states0["GenderPop"].str.split("_", expand=True)
states0[["Males", "Females"]] = states0["GenderPop"].str.split("_", expand=True)
```

```python
states0
```

```python
states0.Males = states0.Males.str.replace('M', '').astype(int)
states0.Females = states0.Females.str.replace('F', '').astype(int)
```

```python
states0
```

```python
states0.info()
```

```python
states0.drop('GenderPop', axis=1, inplace=True)
```
