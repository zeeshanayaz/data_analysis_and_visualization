Linear Euqation
- Point Slope Form
     - y - y1 = m(x - x1) >>> (x1, y1), m
- Two Point Form
     - (y2 - y1)/(y - y1) = (x2 - x1)/(x - x1)  >>> (x1, y1), (x2, y2) 
- Slope Intercept
     - y = mx + c
- Two Intercept Form
     - x/a + y/b c

## Linear Regression
A linear regression model is stats/ML that finds the relationship between dependent and independents features.

| X (Input) | Y (Output) |
|-----------|------------|
|    0      |     2      |
|    1      |     4      |
|    2      |     6      |
|    3      |     8      |
|   -4      |    -6      |


Formula -> y = mx + b

m = Slope

m = tanθ

m = y/x

m = rise / run

gradient

b = intercept

in linear line equation

m = (y2 - y1) / (x2 - x1)

Slope between any point in straight line  is same.

after calculation, m = 2

and by solving on any point with m = 2, y will be 2

i.e:

y = mx + b

4 = 2(1) + b

b = 2

here after calculation

the model will be,

y = mx + y

y = 2x + 2

Since this euqation satisfied all the points so the line will be **ideal fit line**

m = W = Weight

y = B = Bias

### Simple Linear Regression Model

```python
X = data.Height
Y = data.Weight

X , Y
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```
