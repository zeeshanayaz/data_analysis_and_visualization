```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

#### Step1: Data Load

```python
data = pd.read_csv("advertising.csv")
```

```python
data.shape
```

```python
data.head()
```

#### Step2: Separate input data and output
       Data and Labels 

```python
X = data[['TV', 'Radio', 'Newspaper']]
y = data[['Sales']]
```

```python
X
```

```python
y
```

#### Step3: Split Data Training data and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test   = train_test_split(X,y,random_state=1234, test_size=.20)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

#### Step:4 Model Selection

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

#### Step:5 Model Training (Fit Karna)

```python
# supervised learning data+ans
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

#### Step6: Prediction on Training Data

```python
y_pred_train = model.predict(X_train)
```

```python
y_train.values.reshape(160,)
```

```python
y_pred_train.reshape(160,)
```

```python
pd.DataFrame({"y_train":y_train.values.reshape(160,), "y_pred":y_pred_train.reshape(160,) })
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

R-Squared values range from 0 to 1. An R-Squared value of 0 means that the model explains or predicts 0% of the relationship between the dependent and independent variables. A value of 1 indicates that the model predicts 100% of the relationship, and a value of 0.5 indicates that the model predicts 50%, and so on.

```python
r2_score(y_train,y_pred_train)
```

```python
mean_absolute_error(y_train,y_pred_train)
```

#### Step7: Prediction on Test  Data 

```python
y_pred_test = model.predict(X_test)
```

```python
r2_score(y_test,y_pred_test)
```

```python
mean_absolute_error(y_test,y_pred_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#scaler.fit(X)
scaled_x = scaler.fit_transform(X)
columns = scaler.feature_names_in_
```
